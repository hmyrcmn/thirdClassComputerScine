```python
from Bio.Blast import NCBIXML

result_handle = NCBIWWW.qblast("blastn", "nt", "8332116")

blast_records = NCBIXML.parse(result_handle)

for record in blast_records:
    for alignment in record.alignments:
        for hsp in alignment.hsps:
            print("****Alignment****")
            print("sequence:", alignment.title)
            print("length:", alignment.length)
            print("e value:", hsp.expect)
            print(hsp.query[0:75] + "...")
            print(hsp.match[0:75] + "...")
            print(hsp.sbjct[0:75] + "...")

```

```python
from Bio.Blast import NCBIWWW

fasta_string = open("/content/m_cold.fasta").read()
result_handle = NCBIWWW.qblast("blastn", "nt", fasta_string)

# Elde edilen sonucu yazdır
with open("blast_results.xml", "w") as out_handle:
    out_handle.write(result_handle.read())
    
result_handle.close()
```

```python
import xml.etree.ElementTree as ET

# XML dosyasının yolunu belirleyin
file_path = "/content/blast_results.xml"

# XML dosyasını açın
tree = ET.parse(file_path)

# XML ağacının kök düğümüne erişin
root = tree.getroot()

# Tüm etiketleri ve metin içeriğini yazdırın
for elem in root.iter():
    print(elem.tag, elem.text)
```
